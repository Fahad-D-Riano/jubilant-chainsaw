```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
diabetes_data = {
    'x1' : [0.7, 0.8, 0.8, 1.2, 0.6, 1.3, 0.9, 0.9],
    'x2' : [0.7, 0.9, 0.25, 0.8, 0.4, 0.5,0.5, 1.1],
    'y'  : [0, 1, 0, 1, 0, 1, 0, 1]
}
df = pd.DataFrame(diabetes_data)
```

```python
df
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
```

```python

epochs = 10
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df[['x1','x2']], df['y'],test_size = 0.25)
```

```python
def gradient_descent(x1, x2, y, epochs):
    w1 = w2 = 1
    bias = 0
    rate = 0.01
    n = len(x1)
    for i in range(epochs):
        weighted_sum = w1 * x1 + w2 * x2+ bias
        y_predicted = sigmoid_numpy(weighted_sum)

        w1d = (1/n)*np.dot(np.transpose(x1),(y_predicted-y)) 
        w2d = (1/n)*np.dot(np.transpose(x2),(y_predicted-y)) 

        bias_d = np.mean(y_predicted-y)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d

        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}')


    return w1, w2, bias
```

```python
w1, w2, b =gradient_descent(X_train['x1'],X_train['x2'],Y_train,100)
```

```python
def predict(x1,x2,y):
    weighted_sum = w1 * x1 + w2 * x2+ b
    y_predicted = sigmoid_numpy(weighted_sum)
    return y_predicted, y
```

```python
predict(X_test['x1'], X_test['x2'], Y_test)
```
